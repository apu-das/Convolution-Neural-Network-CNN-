```python
img = image.load_img('train/cat.jpg' , target_size=(200,200))
plt.imshow(img)
```

```python
datagen = ImageDataGenerator(
    rotation_range =30,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip  = True,
    vertical_flip = True,
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    fill_mode = 'constant'
)
```

```python
img = image.img_to_array(img)
```

```python
type(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1,200,200,3)
```

```python
i=0
for input in datagen.flow(input_batch, batch_size=1, save_to_dir='aug'):
    i=i+1
    if i==10:
        break
```
